```python
import commons
```

```python
m = "Attack to night"
```

```python
m = commons.preprocessing(m)
```

```python
m
```

```python
def encryptLetter(letter):
    try:
        return key[letter]
    except:
        return None
```

```python
def decryptLetter(code):
    for i in key:
        if key[i] == code:
            return i
```

```python
plaintext = commons.lettersOfPlaintext(m)
```

```python
#plaintext
```

# Encryption

```python
ciphertext = []
for letter in plaintext:
    ciphertext.append(encryptLetter(letter))
```

```python
ciphertext
```

# Decryption

```python
restored_text = []
for i in ciphertext:
    restored_text.append(decryptLetter(i))
```

```python
restored_text
```
